```python
plt.plot(df['index'],df['V Kohli'], linestyle = 'dashdot', linewidth = 5, marker = 'o', markersize = 10)
plt.plot(df['index'],df['RG Sharma'],linestyle =  'dashdot', marker = '>')
plt.title('Kohli Vs Rohit Runs')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.show()
```

```python
plt.plot(df['index'],df['V Kohli'], linestyle = 'dashdot', linewidth = 5, marker = 'o', markersize = 10, label = 'Virat')
plt.plot(df['index'],df['RG Sharma'],linestyle =  'dashdot', marker = '>', label = 'Rohit')
plt.title('Kohli Vs Rohit Runs')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.legend()
plt.show()
```

```python
plt.plot(df['index'],df['V Kohli'], linewidth = 5, marker = 'o', markersize = 10, label = 'Virat')
plt.plot(df['index'],df['RG Sharma'], marker = '>', label = 'Rohit',markersize = 10)
plt.title('Kohli Vs Rohit Runs')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.legend(loc = 'upper left')
plt.show()
```

```python
#Side by side
plt.bar(np.arange(df.shape[0])-0.3, df['2015'], width= 0.3)
plt.bar(np.arange(df.shape[0]), df['2016'], width= 0.3)
plt.bar(np.arange(df.shape[0])+0.3, df['2017'], width= 0.3)
# plt.show()
```

```python
#Side by side
plt.bar(np.arange(df.shape[0])-0.3, df['2015'], width= 0.3)
plt.bar(np.arange(df.shape[0]), df['2016'], width= 0.3)
plt.bar(np.arange(df.shape[0])+0.3, df['2017'], width= 0.3)
plt.xticks(np.arange(df.shape[0]),df['batsman'])
```

```python
#Side by side
plt.bar(np.arange(df.shape[0])-0.3, df['2015'], width= 0.3, color = '#00B6B3')
plt.bar(np.arange(df.shape[0]), df['2016'], width= 0.3, color = '#0056D2')
plt.bar(np.arange(df.shape[0])+0.3, df['2017'], width= 0.3, color = '#22B1F9')
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

### *Stacked bar Chart*

```python
#Side by side
plt.bar(df['batsman'], df['2015'], width= 0.8, color = '#00B6B3', label = '2015')
plt.bar(df['batsman'], df['2016'], width= 0.8,bottom=df['2015'],  color = '#0056D2', label = '2016')
plt.bar(df['batsman'], df['2017'], width= 0.8,bottom = df['2015'] + df['2016'], color = '#22B1F9',label = '2017')
plt.legend()
plt.show()
```

### *Pie Chart*

```python
tips2  = tips.groupby('day')['total_bill'].sum()
```

```python
plt.pie(tips2, labels = tips2.index, autopct= '%0.2f%%')
plt.show()
```

```python
#Explode Shadow
plt.pie(tips2, labels = tips2.index, autopct= '%0.2f%%', explode = [0,0,0,0.05])
plt.show()
```

```python
#Explode Shadow
plt.pie(tips2, labels = tips2.index, autopct= '%0.2f%%', explode = [0,0,0.06,0.05], shadow= True)
plt.savefig('sample.jpeg')
#Show doesn't allow to save
```
