## *Facet Grid*

```python
g = sns.FacetGrid(tips, col="time",  row="sex")
g.map(sns.scatterplot, "total_bill", "tip")
```

```python
g = sns.FacetGrid(tips, col="time",  row="sex")
g.map(sns.histplot, "total_bill")
```

```python
g = sns.FacetGrid(tips, col="time", row="sex")
g.map_dataframe(sns.histplot, x="total_bill", binwidth=2, binrange=(0, 60))
```

```python
g = sns.FacetGrid(tips, col="time", hue="sex")
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
g.add_legend()
```

```python
g = sns.FacetGrid(tips, col="sex", row="time", margin_titles=True, despine=False)
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
g.figure.subplots_adjust(wspace=0, hspace=0)
for (row_val, col_val), ax in g.axes_dict.items():
    if row_val == "Lunch" and col_val == "Female":
        ax.set_facecolor(".95")
    else:
        ax.set_facecolor((0, 0, 0, 0))
```

## *Pair plot*

```python
sns.pairplot(tips)
```

```python
sns.pairplot(iris, hue = 'species')
```

```python
g =sns.PairGrid(data = iris)
g.map(sns.scatterplot)
```

```python
g =sns.PairGrid(data = iris, hue = 'species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
# map_diag -> map_upper -> map_lower
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# vars
g = sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

### JointGrid Vs Jointplot

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='hist',hue='sex')
```

```python
g = sns.JointGrid(data=tips,x='total_bill',y='tip')
g.plot(sns.kdeplot,sns.violinplot)
```
