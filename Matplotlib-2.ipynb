```python
plt.figure(figsize=(12,8))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s = sample_df['runs']*0.3)
#To show annot
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

```python
plt.figure(figsize=(12,8))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s = sample_df['runs']*0.3)
plt.axhline(130, color = 'red')
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

```python
plt.figure(figsize=(12,8))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s = sample_df['runs']*0.3)
plt.axhline(130, color = 'red')
plt.axvline(30, color = 'green')
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

```python
plt.figure(figsize=(12,8))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s = sample_df['runs']*0.3)
plt.axhline(110, color = '#5E765A')
plt.axhline(140, color = 'red')
plt.axvline(30, color = '#5E765A')
plt.axvline(40, color = 'green')
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

```python
fig, ax = plt.subplots(nrows= 2 , ncols= 2,  figsize=(12,8)) #Defining row and columns

ax[0,0].scatter(df['avg'], df['strike_rate'], c = '#0A4101')
ax[0,1].scatter(df['avg'], df['runs'])

ax[1,0].hist(df['runs'])
ax[1,1].hist(df['avg'])
#Title
ax[0,0].set_title('Batsman Average and Strike Rate', c = '#00084D')
ax[0,0].set_ylabel('Strike Rate',c = '#00084D')
ax[0,0].set_xlabel('Average',c = '#00084D')


ax[0,1].set_ylabel('Strike Rate',c = '#00084D')
ax[0,1].set_xlabel('Runs',c = '#00084D')

fig.show()
```

### *3D Surface Plots*

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.log(xx)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
